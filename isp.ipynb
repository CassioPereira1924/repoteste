```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep=';', encoding= 'latin1')
df.head()
```

```python
regiao_metropolitana = ['Rio de Janeiro','Niterói','São Gonçalo','Duque de Caxias','Nova Iguaçu','Belford Roxo','Magé','Queimados','Itaguaí','Seropédica', 'Paracambi','Guapimirim', 'Tanguá', 'Rio Bonito', 'Silva Jardim', 'Araruama' , 'Cabo Frio', 'Búzios', 'São Pedro da Aldeia','Arraial do Cabo', 'Macaé', 'Casimiro de Abreu','São Jão de Meriti']
```

```python
df_metropolitano = df.loc[(df['munic'].isin(regiao_metropolitana)) & (df['ano'] == 2024)]


df_metropolitano.head()
```

```python
df_furto = df_metropolitano.groupby('munic')['furto_celular'].sum().reset_index()
df_furto = df_furto.sort_values(by= 'furto_celular', ascending= False)
df_furto
```

```python
array_furto_celular = np.array(df_furto['furto_celular'])
```

```python
media_furto = np.mean(array_furto_celular)
mediana_furto = np.median(array_furto_celular)
distancia = (media_furto - mediana_furto) / mediana_furto
```

```python
print(f'A média é {media_furto}\nA mediana é {mediana_furto}\nA disntância entre eles é de {(distancia * 100):.2f} %')
```

```python
q1 = np.percentile(array_furto_celular,25)
q3 = np.percentile(array_furto_celular,75)
```

```python
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
```

## Outiliers de cidades da região metropolitana do Rio de Janeiro com mais furtos de celular

```python
df_furto_outliers = df_furto.loc[df_furto['furto_celular'] >= limite_superior]
df_furto_outliers
```

```python
df_furto_outliers[['munic','furto_celular']]
```

```python
df_furto_outliers.to_csv('outliers.csv', index= False)
```

```python
df_furto_outliers = df_furto.loc[df_furto['furto_celular'] <= limite_inferior]
df_furto_outliers
```
