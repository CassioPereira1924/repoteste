```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caramelo = pd.read_csv('CARAMELO.csv', sep= ';')
caramelo['Valor de Venda'] = caramelo['Valor de Venda'].str.replace(',', '.').astype(float)
```

```python
caramelo_rj = caramelo.loc[caramelo['Estado Loja'] == 'RJ']
caramelo_cidades = caramelo_rj.groupby('Cidade Loja')['Valor de Venda'].sum().reset_index()
caramelo_cidades = caramelo_cidades.sort_values(by='Valor de Venda', ascending= False)
caramelo_cidades
```

```python
cidade_loja = caramelo.groupby('Cidade Loja')['Valor de Venda'].sum().reset_index()
cidade_loja = cidade_loja.sort_values(by='Valor de Venda', ascending= False)
cidade_loja
```

```python
cat_produto = caramelo.groupby('Categoria Produto')['Valor de Venda'].mean().reset_index()
cat_produto = cat_produto.sort_values(by='Valor de Venda', ascending= False)
cat_produto
```

```python
teste = caramelo.groupby('Categoria Produto').size().reset_index(name='quantidade')
teste
```

```python
df_eletronicos = caramelo.loc[caramelo['Categoria Produto'] == 'Eletrônicos']
caramelo_marcas  = df_eletronicos.groupby('Marca')['Valor de Venda'].sum().reset_index()
caramelo_marcas = caramelo_marcas.sort_values(by='Valor de Venda', ascending= False)
caramelo_marcas

```

```python
caramelo_marcas.plot( kind='barh', x= 'Marca', y= 'Valor de Venda')
plt.xlabel('Marcas Produto')
plt.ylabel('Valor de Venda')
plt.title('Valor de venda por marca')
plt.show()
```
