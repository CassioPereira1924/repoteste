```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2_2024 = pd.read_csv('2º_2024.csv', sep= ';')
df1_2024 = pd.read_csv('1º_2024.csv', sep= ';')
df1_2023 = pd.read_csv('1º_2023.csv', sep= ';')
df2_2023 = pd.read_csv('2º_2023.csv', sep= ';')
df2_2021 = pd.read_csv('2º_2021.csv', sep= ';', encoding="latin1" )
```

```python
df = pd.concat([df1_2023,df2_2023,df1_2024,df2_2024])
```

```python
df.info()
```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'], format="%d/%m/%Y")


df['mes'] = df['Data da Coleta'].dt.month
df['Ano'] = df['Data da Coleta'].dt.year
df
```

```python
df.info()
```

```python
df_tempo = df.groupby(['mes','Estado - Sigla'])['Valor de Venda'].mean().reset_index()
df_tempo
```

```python
df_comum = df.loc[df['Produto'] == 'GASOLINA']
df_aditivada = df.loc[df['Produto'] == 'GASOLINA ADITIVADA']
df_comum.head()
```

```python
df_gasolina_comum_media = df_comum.groupby('Produto')['Valor de Venda'].mean().reset_index()
df_gasolina_comum_media
```

```python
df_gasolina_comum_aditivada = df_aditivada.groupby('Produto')['Valor de Venda'].mean().reset_index()
df_gasolina_comum_aditivada
```

```python
df_gasosa = pd.concat([df_gasolina_comum_media, df_gasolina_comum_aditivada ])
df_gasosa
```

```python
regioes  = df_comum.groupby(['Ano', 'Regiao - Sigla'])['Valor de Venda'].mean().reset_index()
regioes
```

```python
bandeira = df_comum.groupby('Bandeira')['Valor de Venda'].mean().reset_index()
bandeira = bandeira.sort_values(by= 'Valor de Venda', ascending= False)
bandeira.head(5)
```
